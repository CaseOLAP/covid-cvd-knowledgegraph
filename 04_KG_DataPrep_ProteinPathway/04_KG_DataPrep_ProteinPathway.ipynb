Combine UniProt IDs from both sets of data into a list. This list will then be converted to a .txt file as input for Reactome's API, which will get the pathways for all the UniProt IDs. For some reason, Reactome did not accept .txt so I just ended up copying and pasting the text file.

```python
ump_data_uids = [i.get('UniProt') for i in uniprot_mesh_pub]
drug_ent_uids = []
for drug in drug_ent:
    for entity in ['targets','enzymes','transporters','carriers']:
        for ent in drug[entity]:
            drug_ent_uids.append(ent['uniprot_id'])
data_uids = list(set(ump_data_uids + drug_ent_uids))
```

```python
with open('data_uniprot.txt','w') as f:
    for uid in data_uids:
        f.write(uid + "\n")
    f.close()
```
